```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#Plotting price range distribution of restaurants
business_df = pd.read_csv('data/business_data.csv',low_memory=False)
business_df['price_range'].value_counts().sort_index().plot(kind='bar')
plt.title('Price Range Distribution')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```

```python
#plotting distribution of business ratings 
business_df['business_rating'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Business Ratings')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()
```

```python
# Creating an interactive geograhic distribution scatter map
fig = px.scatter_mapbox(
    business_df,
    lat='latitude',  
    lon='longitude',  
    color='state',  
    hover_name='city',  
    title='Geographic Distribution of Restaurants',
    color_continuous_scale=px.colors.sequential.Viridis,  
    size_max=15, 
    zoom=3,  
    height=700  
)

# Set the map style
fig.update_layout(mapbox_style='carto-darkmatter')

fig.write_html("business_map.html")

# Show the map
fig.show()
```

```python
review_df = pd.read_csv('data/review_data.csv')
review_df['review_rating'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Review Count by Star Ratings')
plt.xlabel('Star Ratings')
plt.ylabel('Count')
plt.show()
```

```python
user_df = pd.read_csv('data/user_data.csv')
# Apply log transformation to review counts (add 1 to avoid log(0))
user_df['log_review_count'] = np.log1p(user_df['user_review_count'])

# Plot the log-transformed histogram
user_df['log_review_count'].plot(kind='hist', bins=30, color='green', edgecolor='black')
plt.title('Log-Transformed Distribution of User Review Counts')
plt.xlabel('Log(Number of Reviews)')
plt.ylabel('User Count')
plt.show()
```

```python
user_df['average_stars'].plot(kind='hist', bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Average User Ratings')
plt.xlabel('Average Stars')
plt.ylabel('User Count')
plt.show()
```

```python
plt.scatter(user_df['user_review_count'], user_df['average_stars'], alpha=0.5)
plt.title('Correlation Between Review Count and Average Stars')
plt.xlabel('User Review Count')
plt.ylabel('Average Stars')
plt.show()
```

```python
checkin_df = pd.read_csv('data/checkin_data.csv')
checkin_df['log_no_of_checkins'] = np.log1p(checkin_df['no_of_checkins'])
checkin_df['log_no_of_checkins'].plot(kind='hist', bins=10, color='cyan', edgecolor='black')
plt.title('Distribution of Check-in Counts - Log Transformed')
plt.xlabel('Check-in Count (Log)')
plt.ylabel('Business Count')
plt.show()
```
